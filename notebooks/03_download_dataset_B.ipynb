```python
RAW_B_DIR = Path("../raw_audio/dataset_B")
RAW_B_DIR.mkdir(parents=True, exist_ok=True)

downloaded = 0
failed = []

for f in files:
    filename = f["name"]
    try:
        data = supabase.storage.from_("audio-recordings").download(filename)
        save_path = RAW_B_DIR / filename
        with open(save_path, "wb") as out:
            out.write(data)
        downloaded += 1
        print(f"Downloaded: {filename}")
    except Exception as e:
        failed.append(filename)
        print(f"Failed: {filename} — {e}")

print(f"\nDownloaded {downloaded} of {len(files)} files")
if failed:
    print(f"Failed files: {failed}")
```

```python
webm_files = list(RAW_B_DIR.glob("*.webm"))
print(f"Total webm files: {len(webm_files)}")

# Check the truth/lie split by filename
truth_count = len([f for f in webm_files if "_truth" in f.name])
lie_count = len([f for f in webm_files if "_lie" in f.name])
print(f"Truth files: {truth_count}")
print(f"Lie files: {lie_count}")
```
